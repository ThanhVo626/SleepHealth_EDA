```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

```

```python
grouped_data = df.groupby('Gender').agg({'Age': 'mean', 'Sleep Duration': 'mean'})
fig, ax = plt.subplots(figsize=(8, 6))
grouped_data.plot(kind='bar', ax=ax)
ax.set_xlabel('Gender')
ax.set_ylabel('Average Age / Sleep Duration')
ax.set_title('Average Age and Sleep Duration by Gender')
plt.xticks(rotation=0)
plt.show()
```
